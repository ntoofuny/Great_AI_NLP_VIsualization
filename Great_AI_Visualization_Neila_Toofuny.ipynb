```python
# This command might take a little while. 
nlp = spacy.load('en')
#loads english module of spacy
```

```python
ai_raw = open('./input/ai.txt').read()
```

```python
ai_awake = nlp(ai_raw)
#parse war and peace
ai_awake[4]
ai_awake[4].pos_
```

```python
ai_awake[3].is_stop
```

```python
ai_awake = [word for word in ai_awake if word.is_stop== False and word.pos_ != 'PUNCT' and word.pos_ != 'SPACE']
    #only put words in the list that aren't stop words, punctuation and spaces
```

```python
ai_awake[4]
```

```python
count_dic = Counter([word.string.strip().lower() 
                     for word in ai_awake])
count_dic
#get word counts for every word in way and peace
```

```python
print(count_dic[200])
#play around with dictionary, check contents
```

```python
list(count_dic.elements())
#print complete list of elements
```

```python
most_common = count_dic.most_common(1000)
most_common
#get list of 1000 most common key pairs in dictionary
```

```python
most_common[47]
```

```python
ai_sentences = LineSentence('./input/ai.txt') 
#I made an additional version of the list of war_and_peace sentences using the LineSentence() function 
```

```python
ai_model = Word2Vec(ai_sentences, size=100, window=4, min_count=1, workers=4, iter = 10)
#pass in LineSentence() object, works better than simple list of sentences for some reason
#iter = epochs
```

```python
ai_model.most_similar("man")
#check model works with most_similar function
```

```python
#now that we have our model, we can initialize our lost 
top_words_ai = []
for word_pair in most_common:
    if word_pair[0] in ai_model.wv.vocab:
        top_words_ai.append(word_pair[0])

top_words_ai
#yay! now we have a list of the 1000 most common words in WAR AND PEACE!
    
```

```python
#t-SNE visualization
# This assumes words_top_ted is a list of strings, the top 1000 words
words_top_vec_ai = ai_model[top_words_ai]
```

```python
for entry in words_top_vec_ai:
    print(type(entry))
```

```python
print(words_top_vec_ai[0])
#curious about the contents of words_top_vec_wap after input into model
```

```python
from sklearn.manifold import TSNE
#initialize TSNE object
tsne = TSNE(n_components=2, random_state=0)
words_top_ai_tsne = tsne.fit_transform(words_top_vec_ai)
#TSNE uses conditional probabilities, joint probabilities, etc
#reduces the dimension of data to 2-3 dimensions so it can be plotted easily 
```

```python
p = figure(tools="pan,wheel_zoom,reset,save",
           toolbar_location="above",
           title="word2vec T-SNE for most common words")

source = ColumnDataSource(data=dict(x1=words_top_ai_tsne[:,0],
                                    x2=words_top_ai_tsne[:,1],
                                    names=top_words_ai))
#what the hell is this?!

p.scatter(x="x1", y="x2", size=8, source=source)

labels = LabelSet(x="x1", y="x2", text="names", y_offset=6,
                  text_font_size="8pt", text_color="#555555",
                  source=source, text_align='center')
p.add_layout(labels)

show(p)
```
